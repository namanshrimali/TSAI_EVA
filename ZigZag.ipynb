```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('dark_background')
```

```python

def plot_graph(data: tuple, plot_title: str, axis_lables: tuple, save: bool):
    fig, ax = plt.subplots()
    ax.set(xlabel=axis_lables[0], ylabel=axis_lables[1],title='Cyclic Triangle Plot')
    ax.grid()
    ax.plot(data[0], data[1])

    if save:
        fig.savefig(f'{plot_title}.png')
    plt.show() # dunno what it does, graphs plot without it too 
```

```python
import math
def get_triangle_points(x_vals, y_range, step_size):
    y_val = []
    slope = (y_range[1]-y_range[0])/step_size
    for x in x_vals:
        slope_coeff = None
        c = None
        if math.ceil(x/step_size)%2 == 1:
            slope_coeff = 1
            c = y_range[1]
        else:
            slope_coeff = 0
            c = 0
        y = slope * slope_coeff * x%step_size + c
        y_val.append(y)
        
    return np.array(y_val)
        
```

```python
def draw_zig_zag(step_size: int, x_range: tuple, y_range: tuple):
    x_points = np.arange(x_range[0], x_range[1], step_size)
    y_points = get_triangle_points(x_points, y_range, step_size)
    plot_graph((x_points, y_points), 'Cyclic Triangle Plot', ('zag', 'zig'), save=True)

```

```python
draw_zig_zag(2, (0, 20), (0, 10))
```
